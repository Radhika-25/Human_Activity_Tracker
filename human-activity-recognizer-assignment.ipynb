```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier 
```

```python
test = pd.read_csv('/kaggle/input/eda-of-human-activity-recognition/f_test.csv')
```

```python
train = pd.read_csv('/kaggle/input/eda-of-human-activity-recognition/f_train.csv')
train.head()
```

```python
train_data = train.iloc[:,:-1]
train_label = train.iloc[:,-1]
```

# Decision Tree Classifier

```python
classifier = DecisionTreeClassifier()
classifier.fit(train_data,train_label)

from sklearn import tree
plt.figure(figsize=(10,10))
tree.plot_tree(classifier,filled=True)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(classifier, train_data, train_label, cv=10)
```

```python
test.head()
```

```python
test_data = test.iloc[:,:-1]
test_labels = test.iloc[:,-1]
test_pred = classifier.predict(test_data)
```

```python
for i in range(10):
    print('Predicted -',test_pred[i])
    print('Actual -',test_labels[i],'\n')
```

```python
from sklearn.metrics import classification_report,accuracy_score 
print(accuracy_score(test_labels,test_pred))
print(classification_report(test_labels,test_pred))
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(train_data,train_label)
```

```python
test_pred = classifier.predict(test_data)

print(accuracy_score(test_labels,test_pred))
print(classification_report(test_labels,test_pred))
```

```python
for i in range(10):
    print('Predicted -',test_pred[i])
    print('Actual -',test_labels[i],'\n')
```

# SVM Classifier 

```python
from sklearn.svm import SVC
classifier = SVC(kernel='linear')
classifier.fit(train_data,train_label)
```

```python
test_pred = classifier.predict(test_data)

print(accuracy_score(test_labels,test_pred))
print(classification_report(test_labels,test_pred))
```

```python
for i in range(10):
    print('Predicted -',test_pred[i])
    print('Actual -',test_labels[i],'\n')
```

```python
classifier = SVC(kernel='poly',degree=6) #OBSERVATION: As the degree of the polynomial used for the kernel increases,
                                         #the accuracy decreases here 
classifier.fit(train_data,train_label)

test_pred = classifier.predict(test_data)

print(accuracy_score(test_labels,test_pred))
print(classification_report(test_labels,test_pred))
```

```python
classifier = SVC(kernel='sigmoid',gamma='scale') 
classifier.fit(train_data,train_label)

test_pred = classifier.predict(test_data)

print(accuracy_score(test_labels,test_pred))
print(classification_report(test_labels,test_pred))
```

```python
classifier = SVC(kernel='rbf') 
classifier.fit(train_data,train_label)

test_pred = classifier.predict(test_data)

print(accuracy_score(test_labels,test_pred))
print(classification_report(test_labels,test_pred))
```

#### Observations - the best accuracy in SVM method is given by the Linear followed by the rbf kernel. The sigmoid kernel gives the worst accuracy and must not be used
